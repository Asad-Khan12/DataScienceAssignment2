```python
import pandas as pd

def read_worldbank_dataframe(filename):
   # Read the CSV file into a pandas dataframe
    df1 = pd.read_excel(filename)
    
    # Set the index to be the country names
    df_countries=df1['Country name']

    # Transpose the dataframe to create a years-as-columns dataframe
    df_years=df1.T.iloc[6:]
    df_years = df_years.transpose()

    # Create a countries-as-columns dataframe by resetting the index

    # Return the two dataframes
    return df_years, df_countries
```

```python
years_df, countries_df= read_worldbank_dataframe('D:/sheharyar/datascience/climate-change.xls')
```

# Explore the statistical properties of a few indicators

# Statistical Analysis of China

```python

# Define the list of indicators and countries of interest

indicators = ["EN.ATM.CO2E.PP.GD.KD", "AG.LND.EL5M.ZS", "IE.PPI.TRAN.CD"]
countries = ['CHN']

# Read the World Bank data into a pandas dataframe
df=pd.read_excel('D:/sheharyar/datascience/climate-change.xls')
dfd=df

# Filter the dataframe to include only the indicators and countries of interest
df = df[df['Country code'].isin(countries) & df['Series code'].isin(indicators)]

# Pivot the dataframe to create a country-indicator dataframe
df_pivot = df.pivot(index='Country code', columns='Series code' )

# Calculate summary statistics for the country-indicator dataframe
df_stats = df_pivot.describe()

# Print the summary statistics
print(df_stats)
```
